```python
import pandas as pd

# Baca data dari file Excel
df = pd.read_excel('list_senyawa_fix.xlsx')

# Tampilkan data untuk memastikan pembacaan berhasil
df.head(155)
```

```python
from rdkit import Chem
from rdkit.Chem import MACCSkeys

# Inisialisasi list untuk menyimpan fingerprint
fingerprints = []

# Loop melalui setiap baris data
for index, row in df.iterrows():
    smiles = row['Smiles']
    mol = Chem.MolFromSmiles(smiles)

    if mol:
        maccs_keys = MACCSkeys.GenMACCSKeys(mol)
        fingerprint_values = maccs_keys.ToBitString()
        fingerprints.append(fingerprint_values)
    else:
        fingerprints.append(None)

# Tambahkan kolom fingerprint ke DataFrame
df['MACCS Keys Fingerprint'] = fingerprints

# Tampilkan DataFrame yang telah diperbarui
df.head(155)
```

# **Fix Code**

```python
fingerprints = []

for index, row in df.iterrows():
    smiles = row['Smiles']
    mol = Chem.MolFromSmiles(smiles)

    if mol:
        maccs_keys = MACCSkeys.GenMACCSKeys(mol)
        fingerprint_values = maccs_keys.ToBitString()
        fingerprints.append(fingerprint_values)
    else:
        fingerprints.append(None)

df['MACCS Keys Fingerprint'] = fingerprints

for i in range(167):
    bit_column_name = f'bits{i}'
    df[bit_column_name] = df['MACCS Keys Fingerprint'].apply(lambda x: x[i] if x else None)

df.drop(columns=['MACCS Keys Fingerprint'], inplace=True)

df
```

```python
# Menulis DataFrame ke file Excel
df.to_excel('hasil_fingerprint.xlsx', index=False)

# Mengunduh file Excel yang dihasilkan
files.download('hasil_fingerprint.xlsx')
```
