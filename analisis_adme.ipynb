```python
!pip install rdkit-pypi # Install the rdkit package. This line needs to be run first.
import pandas as pd
import glob
from rdkit import Chem # This will now work if the above installation was successful.
from rdkit.Chem import Descriptors
from rdkit.Chem import Crippen

# Menggunakan glob untuk mendapatkan semua file CSV dalam direktori
file_pattern = "chunk_20.csv"  # Ganti dengan pola nama file yang sesuai
all_files = glob.glob(file_pattern)

# Membaca dan menggabungkan semua file menjadi satu DataFrame
dfs = [pd.read_csv(file) for file in all_files]
full_df = pd.concat(dfs, ignore_index=True)

# Tampilkan beberapa baris dari DataFrame gabungan
print(full_df.head())
```

```python
# Misalkan kolom SMILES ada di 'Canonical SMILES'
full_df['Molecule'] = full_df['canonical_smiles'].apply(Chem.MolFromSmiles)

# Hapus baris yang tidak valid
full_df = full_df[full_df['Molecule'].notnull()]
```

```python
# Hitung sifat ADME
full_df['LogP'] = full_df['Molecule'].apply(lambda mol: Crippen.MolLogP(mol))
full_df['Molecular Weight'] = full_df['Molecule'].apply(lambda mol: Descriptors.MolWt(mol))
full_df['Num H Donors'] = full_df['Molecule'].apply(lambda mol: Descriptors.NumHDonors(mol))
full_df['Num H Acceptors'] = full_df['Molecule'].apply(lambda mol: Descriptors.NumHAcceptors(mol))
full_df['Polar Surface Area'] = full_df['Molecule'].apply(lambda mol: Descriptors.TPSA(mol))
```

```python
print(full_df[['canonical_smiles', 'LogP', 'Molecular Weight', 'Num H Donors', 'Num H Acceptors', 'Polar Surface Area']])
```

```python
import pandas as pd
```

```python
# Jika file hasil ADME dalam bentuk Excel
df = pd.read_excel('ADME_analysis_results_chunk_20.xlsx')

# Tampilkan beberapa baris pertama
print(df.head())
```

```python
# Tentukan kriteria seleksi
logp_condition = (df['LogP'] > 0) & (df['LogP'] < 5)
mw_condition = (df['Molecular Weight'] > 300) & (df['Molecular Weight'] < 500)
h_donors_condition = (df['Num H Donors'] >= 0) & (df['Num H Donors'] <= 5)
h_acceptors_condition = (df['Num H Acceptors'] >= 0) & (df['Num H Acceptors'] <= 10)
psa_condition = df['Polar Surface Area'] < 140

# Gabungkan semua kondisi
criteria = logp_condition & mw_condition & h_donors_condition & h_acceptors_condition & psa_condition

# Seleksi DataFrame berdasarkan kriteria
selected_df = df[criteria]

# Tampilkan hasil seleksi
print(selected_df[['canonical_smiles', 'LogP', 'Molecular Weight', 'Num H Donors', 'Num H Acceptors', 'Polar Surface Area']])
```

```python
# Simpan hasil seleksi ke file Excel
selected_df.to_excel('ADME_selected_results_chunk_20.xlsx', index=False, engine='openpyxl')

# Download file Excel (jika di Colab)
from google.colab import files
files.download('ADME_selected_results_chunk_20.xlsx')
```
